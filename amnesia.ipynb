Variable-order Markov chains (VOMC)
---

We first define a class for a context tree, with methods which shorten memory (pruning). Specific methods shorten memory for reversible and non-reversible models. Reversibility requires that 

- if `seq` is a context, then so is its inverse, and
- the *amnesia* property, i.e. if `seq` is a context, then any sequence with `seq` as a prefix is a context.

In the non-reversible case, we impose only the second property even though it isn't required by the model as in the reversible case.  

```python
import numpy as np
from matplotlib import pyplot as plt
import copy
```

```python
class Tree(object):
    def __init__(self, k, memoryLength=0):
        self.previous = None
        self.k = k
        for i in range(memoryLength):
            self.ExtendAll()
        
    def ExtendAll(self):
        """
        Extends tree by 1 in all leaves
        """
        if self.previous == None:
            self.previous = { i : Tree(self.k) for i in range(self.k) }
        else:
            for key,tree in self.previous.iteritems():
                tree.ExtendAll()
                
    def MakeContext(self,seq):
        """
        Truncates tree such that seq is a context
        """
        if len(seq) == 0:
            self.previous = None
        elif not self.previous == None:
            self.previous[seq[-1]].MakeContext(seq[:-1]) 
            for i in range(self.k):
                self.previous[i].MakeContext(seq)
        
    def MakeReversibleContext(self,seq):
        """
        Truncates tree such that seq is a context
        with the constraints of reversibility
        """
        self.MakeContext(seq)
        seqReverse = list(seq)
        seqReverse.reverse()
        self.MakeContext(seqReverse)
        
    def FindContext(self,seq):
        """
        For the given sequence, find the longest context
        """
        if self.previous == None:
            return []
        else:
            return self.previous[seq[-1]].FindContext(seq[:-1]) + seq[-1:]
        
    def NotIn(self,contextDict,i,after=[]):
        """
        Returns a list of all sequences of length i of the form:
        before + after
        which are not a key of `contextDict`, and such that `before` is not already a context in the tree
        """
        if self.previous == None:
            return []
        if len(after)==i:
            if not contextDict.has_key(tuple(after)):
                return [after]
            else:
                return []
        out = []
        for x in range(self.k):
            out += self.previous[x].NotIn(contextDict,i,[x]+after)
        return out
    
    def Print(self,last=[]):
        """
        Print a nice visualization of the graph with a prepend determined by `last`
        """
        prepend = [ '    ' if last[j] else '|   ' for j in range(len(last))]
        if self.previous == None:
            print ''.join(prepend)
        else:
            for x in range(self.k):
                print ''.join(prepend+['|-> '+str(x)])
                self.previous[x].Print(last=last+[x==self.k-1])
            
```

Now, we define classses for Bayesian inference in variable-order Markov chains. The second class inherits from the first and implements palindrome reinforcement for reversible variable-order Markov chain.

```python
class VOMC:
    """
    Represents a variable order Markov chain with reinforcement.
    The prior weights are the weights on a de Bruijn sequence, with priorWeight per transition
    """
    
    def __init__(self,tree,maxMemory,priorWeight):
        self.tree = tree                # the context tree
        self.maxMemory = maxMemory      # maximum length of memory
        self.priorWeight = priorWeight  # weight of de Bruijn sequence used to define prior
        self.k = tree.k                 # number of states
        self.weights = {}               # reinforcement weights
    
    def PriorWeight(self,context):
        """
        Outputs the total weight out of a context in the prior
        """
        return self.priorWeight*(self.k**(self.maxMemory-len(context)))
    
    def UpdateWithSequence(self,seq):
        """
        Given an observed sequence, update weights (from a palindrome that joins the reverse
        and the sequence). 
        """
        # Update with each transition
        i = self.maxMemory
        while i <= len(seq)-1:
            self.AddTransition(self.tree.FindContext(seq[i-self.maxMemory:i]),seq[i])
            i += 1
        
    def AddTransition(self,context,x):
        """
        Increase weights by one transition
        """
        context = tuple(context)
        if not self.weights.has_key(context):
            self.weights[context] = dict.fromkeys(np.arange(self.k),0)
            self.weights[context]['total'] = 0
        self.weights[context][x] += 1
        self.weights[context]['total'] += 1
             
    def LogTransitionProbability(self,context,x):
        """
        Outputs the log of the probability of a transition
        """
        weightTotal = self.PriorWeight(tuple(context))
        weight = weightTotal/float(self.k)
        if self.weights.has_key(tuple(context)):
            weight += self.weights[tuple(context)][x]
            weightTotal += self.weights[tuple(context)]['total']
        return np.log(weight/weightTotal)
    
    def LogProbability(self,seq):
        """
        Outputs the log transition probability of a sequence. 
        The sequence must start from the last step of the sequence used to update
        the object. If the object has not been updated with a sequence (self.last
        doesn't exist), this method returns the prior log probability of the sequence.
        """
        # Compute probability for every step and update
        logP = 0
        i = self.maxMemory
        while i <= len(seq)-1:
            context = self.tree.FindContext(seq[i-self.maxMemory:i])
            x = seq[i]
            logP += self.LogTransitionProbability(context,x)   # Add the log-probability of this step
            self.AddTransition(context,x)                      # Reinforce the transition
            i += 1
        self.last = seq[-self.maxMemory:]
        return logP
        
```

```python
class RevVOMC(VOMC):
    """
    Represents a reversible, variable order Markov chain with reinforcement.
    The prior weights are the weights on a de Bruijn sequence, with priorWeight per transition
    """
    def UpdateWithSequence(self,seq):
        """
        Given an observed sequence, update weights (from a palindrome that joins the reverse
        and the sequence). 
        """
        # Define the palindrome
        palindrome = copy.copy(list(seq))
        palindrome.reverse()
        palindrome.extend(list(seq))
        # Update with each transition
        i = self.maxMemory
        while i <= len(palindrome)-1:
            self.AddTransition(self.tree.FindContext(palindrome[i-self.maxMemory:i]),palindrome[i])
            i += 1
        # Set the last state to the final state of the sequence
        self.last = palindrome[-self.maxMemory:]

    def LogProbability(self,seq):
        """
        Outputs the log transition probability of a sequence. 
        The sequence must start from the last step of the sequence used to update
        the object. If the object has not been updated with a sequence (self.last
        doesn't exist), this method returns the prior log probability of the sequence.
        """
        try:
            if not seq[:self.maxMemory] == self.last:
                print "Last state doesn't match the final state of this object!"
                return 
        except AttributeError:
            #print "Outputting prior probability"
            self.UpdateWithSequence(seq[:self.maxMemory])    # In this case, we assume we want the prior
                                                             # probability of the sequence.
        # Compute probability for every step and update
        logP = 0
        i = self.maxMemory
        while i <= len(seq)-1:
            context = self.tree.FindContext(seq[i-self.maxMemory:i])
            x = seq[i]
            logP += self.LogTransitionProbability(context,x)   # Add the log-probability of this step
            self.AddTransition(context,x)                      # Reinforce the forward transition
            context = copy.copy(seq[i-self.maxMemory:i+1])
            context.reverse()
            x = context[-1]
            context = self.tree.FindContext(context[:-1])
            self.AddTransition(context,x)                      # ... and the backward transition
            i += 1
        self.last = seq[-self.maxMemory:]
        return logP
    
```

This illustrates how to build a context tree and a reversible VOMC model.

```python
t = Tree(5,3)
t.MakeReversibleContext([1])
t.MakeReversibleContext([3,2])
t.MakeReversibleContext([0,1])
t.Print()
```

```python
r = RevVOMC(t,4,0.1)
r.LogTransitionProbability([3,2],1)
r.UpdateWithSequence([2,1,3,4,4,4,3,3,2,3,3,2,2,3,3,3,2,3,2,2,2,2])
r.LogProbability([2,2,2,2,3,2,3,2,2,3,3,2,4,4,1,4,1])
```

Model selection procedure for VOMC
----

1. *Initialize*: Start from model with the maximum possible length of memory $n$, with prior parametrized by `priorWeight`. 
2. *Trim data-poor contexts*: For every leaf, if there are not more than 5 occurrences of the context, shorten memory. Recur until the tree doesn't change.
3. *Trim where there is no evidence of memory*: Starting from the longest contexts, compare the log likelihood of the current model with that of a model with the context trimmed. Remove if Bayes factor is smaller than `threshold`. 

The parameters `priorWeight` and `threshold` are optimized via cross-validation on the training set.

We define two classes, `ModelSelectionVOMC`, which implements the procedure above, and a subclass `RevModelSelectionVOMC`, which implements the procedure with a reversible model. 

```python
class ModelSelectionVOMC:
    """
    Implements a Bayesian model selection procedure for a Variable-Order Markov chain
    """
    def __init__(self,k,maxMemory,priorWeight,threshold):
        self.k = k
        self.maxMemory = maxMemory
        self.priorWeight = priorWeight
        self.threshold = threshold
        self.tree = Tree(k,maxMemory)
        self.minOccur = 5            # Minimum number of occurences per context
    
    def MakeContext(self,context):
        """
        Makes input sequence a context in the model.
        """
        self.tree.MakeContext(context)
    
    def TrimDataPoor(self,seq):
        """
        Trims the context tree until all leaves have at least `self.minOccur` occurrences in `seq`
        """
        for i in range(self.maxMemory):
            # Make a dictionary with all sequences of length i that appear in `seq` more than `self.minOccur` times
            subseqs = [seq[j:j+i] for j in xrange(len(seq)-i)]
            occur = {}
            for subseq in subseqs:
                if subseqs.count(subseq) >= self.minOccur:
                        occur[tuple(subseq)] = None
            # For each branch of length i in the tree, trim if the branch is not in `occur`
            for context in self.tree.NotIn(occur,i):
                self.MakeContext(context)
                
    def Trim(self,seq):
        """
        Trim the context tree doing iterative Bayes factor comparisons
        """
        i = self.maxMemory - 1
        while i >= 0:
            untrimmed = self.tree.NotIn({},i)
            for potentialContext in untrimmed:
                self.UpdateModel(potentialContext,seq)
            i += -1
    
    def BuildModel(self):
        """
        Outputs a VOMC object with the current `self.tree`
        """
        tree = copy.deepcopy(self.tree)
        return VOMC(tree,self.maxMemory,self.priorWeight)
    
    def UpdateModel(self,potentialContext,seq):
        """
        Tests whether trimming the tree at `potentialContext` is good using the data in `seq`,
        and updates the model accordingly.
        """
        modelCurrent = self.BuildModel()
        currentTree = copy.deepcopy(self.tree)
        self.MakeContext(potentialContext)
        modelTrimmed = self.BuildModel()
        currentLogP = modelCurrent.LogProbability(seq)
        trimmedLogP = modelTrimmed.LogProbability(seq)
        print "Attempting to trim at "+str(potentialContext)
        print "     Bayes Factor "+str(trimmedLogP-currentLogP)
        if trimmedLogP-currentLogP >= self.threshold:
            print "     Accepted!"
        else:
            self.tree = currentTree
            
    def ModelSelection(self,seq):
        """
        Outputs the VOMC object selected using `seq`
        """
        self.TrimDataPoor(seq)
        print "Context tree after trimming data poor contexts:"
        self.tree.Print()
        self.Trim(seq)
        print "\n\nContext tree after trimming:"
        self.tree.Print()
        return self.BuildModel()
    
```

```python
class RevModelSelectionVOMC(ModelSelectionVOMC):
    """
    Specializes ModelSelectionVOMC to reversible processes.
    """
    
    def MakeContext(self,context):
        """
        Makes a reversible context
        """
        self.tree.MakeReversibleContext(context)
    
    def BuildModel(self):
        """
        Outputs a RevVOMC object with the current `self.tree`
        """
        tree = copy.deepcopy(self.tree)
        return RevVOMC(tree,self.maxMemory,self.priorWeight)
    
```

Here, we demonstrate an example of model selection with an artificial sequence.

```python
seq = [2,1,3,4,4,4,3,3,2,3,3,2,2,3,3,3,2,3,2,2,2,2,1,2,1,2,2,2,1,2,1,4,3,2,2,3,4,3,4,3,4,3,4,3,3,3,3,3,3,3,3,3,3,3,4,3,4,3,4,3,4,3,4,3,2,2,2]
a = RevModelSelectionVOMC(5,4,0.1,0)
model = a.ModelSelection(seq)         # select a model using the sequence seq
model.UpdateWithSequence(seq)         # update the model with the sequence seq
```

```python
# predict the probability of a new sequence with the selected model
newSeq = [3,2,2,2,2,2,3,4,1,2,1,1,1,2,3,3,3,2,3,2,3,2,2,2,4]
model.LogProbability(newSeq)
```
